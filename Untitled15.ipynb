```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
credit_card_data = pd.read_csv('/content/creditcard.csv')
```

separating the data for analysis

```python
legit = credit_card_data[credit_card_data.Class == 0]
fraud = credit_card_data[credit_card_data.Class == 1]
```

statistical measures of the data

```python
legit.Amount.describe()
```

#under-sampling

build a sample dataset containing similar distribuation of normal transaction and fraudulent transactions

Number of fraudulent Transaction-->492




```python
legit_sample = legit.sample(n=492)
```

Concatenating two Dataframes

```python
new_dataset = pd.concat([legit_sample, fraud], axis=0)
```

```python
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
new_dataset['Class'].value_counts
```

```python
new_dataset.groupby('Class').mean()
```

Splitting the data info Features & Targets

```python
X = new_dataset.drop(columns='Class', axis=1)
Y = new_dataset['Class']
```

```python
print(X)
```

```python
print(Y)
```

Split the data info Training data & Testing data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.2, stratify=Y, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

MODEL TRAINING

Logistic Regression

```python
model = LogisticRegression()
```

training the logistic regression model with training data

```python
model.fit(X_train, Y_train)
```

MODEL EVALUATION

Accuracy Score

accuracy on training data

```python
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print('Accuracy on trainig data : ', training_data_accuracy)
```

#accuracy on test data

```python
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print('Accuracy score on Test Datab : ', test_data_accuracy)
```
